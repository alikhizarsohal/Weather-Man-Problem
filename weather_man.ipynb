```python
import os
import csv

def extract_column_data(file_path, column_to_extract,column_pkt):
    try:
        with open(file_path, mode='r') as file:
            csv_reader = csv.DictReader(file)
            # Check if the column name exists in the CSV file
            if column_to_extract not in csv_reader.fieldnames:
                raise ValueError(f"Column '{column_to_extract}' not found in the CSV file.")
            
            # Use a generator to yield the specific column values
            # yield max(row[column_to_extract] for row in csv_reader)
            print("--------------------------------------")
            for row in csv_reader:
                row_list = list(row.values())
                yield [row[column_to_extract],row_list[0]]
            
    except FileNotFoundError:
        print(f"File not found at path: {file_path}")
    except ValueError as ve:
        print(f"Error: {ve}")
    except Exception as e:
        print(f"An error occurred: {str(e)}")

folder_name = 'Dubai_weather'
folder_path = f'dataset/{folder_name}/'

year = '2004'
# Desired file name pattern or substring to match
desired_filename_substring = year  # Example: files containing 'Dubai_weather_2004'
column_to_extract = 'Max TemperatureC'  # Column name to extract
column_pkt = 'GST'  # Packet column name to extract
# List to store extracted data
extracted_data = []
temp_date = []
pkt_date = None
max_temperature = float('-inf')
try:
    # Iterate over files in the directory
    for filename in os.listdir(folder_path):
        # Check if the filename matches the desired pattern
        if desired_filename_substring in filename:
            print("Filename:", filename)
            file_path = os.path.join(folder_path, filename)
            # Extract the column data using the generator
            data_gen = extract_column_data(file_path, column_to_extract, column_pkt)
            # Display the extracted data if data_gen is not None
            if data_gen:
                i=0
                for row in enumerate(data_gen):
                    if row[1][0] == '' or row[1][1] == '':
                        continue
                    value = float(row[1][0])
                    pkt = row[1][1]
                    if value > max_temperature:
                        max_temperature = value
                        temp_date = filename
                        pkt_date = pkt
                        max_file = filename
                    

except FileNotFoundError:
    print(f"Directory not found: {folder_path}")

except Exception as e:
    print(f"An error occurred: {str(e)}")

print("Max temperature:", max_temperature)
print("Date:", pkt_date)
print("File:", max_file)
```
